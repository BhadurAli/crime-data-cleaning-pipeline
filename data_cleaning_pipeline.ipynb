## Handling Missing Values

Removed the Latitide, Longitude and Incident_datetime column because I cannot fill them other wise it will give wrong infromation about place and time of insident. However I have another way to drop latitude and longitude column because I have district, city state and address column

```python
df.columns
```

```python
df.dropna(subset=["latitude","longitude","incident_datetime"], inplace= True)
```

```python
df.shape
```

```python
df.isnull().sum()/len(df)*100
```

Fill "suspect_id" with "SUS00000" for consistency, "suspect_first_name" with  "UNKNOWN" and "suspect_last_name" with "UNKNOWN""

```python
df.fillna({
    "suspect_id" : "SUS00000",
    "suspect_first_name" : "UNKNOWN",
    "suspect_last_name" : "UNKNOWN"
}, inplace= True)
```

### SUSPECT AGE, SUSPECT GENDER, SUSPECT RACE

Again filled "suspect_age" with median of its age, "suspect_gender" : with mode of gender and "suspect_race" with "Unknown"

```python
df.fillna({
    "suspect_age" : df["suspect_age"].median(),
    "suspect_gender" : df["suspect_gender"].mode()[0],
    "suspect_race" : "Unknown"
}, inplace = True)
```

### VICTIM ID, FIRST NAME, LAST NAME, AGE, GENDER

```python
df.fillna({
    "victim_id" : "VIC00000",
    "victim_first_name" : "UNKNOWN",
    "victim_last_name" : "UNKNOWN",
    "victim_age" : df["victim_age"].median(),
    "victim_gender" : "UNKNOWN",
    "victim_phone" : "000-000-0000"
    
}, inplace= True)
```

```python
df.fillna({
    "weapon_used" : "Unknown"
}, inplace = True)
```

```python
df["severity"]=df["severity"].fillna("Unknown")
```

```python
df["num_arrests"].unique()
```

```python
df["case_status"]=df["case_status"].fillna("No Data")
df["resolution"]=df["resolution"].fillna("Unknown")
df["num_arrests"]=df["num_arrests"].fillna(df["num_arrests"].median())
df["property_loss_usd"]=df["property_loss_usd"].fillna("0")
df["reported_online"]=df["reported_online"].fillna(df["reported_online"].mode()[0])
df["notes"]=df["notes"].fillna("No Available Notes")
df["badge_number"]=df["badge_number"].fillna(df["badge_number"].median())
```

```python
df.isnull().sum()/len(df)*100
```
